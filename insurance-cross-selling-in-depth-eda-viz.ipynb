# <b><span style='color:#45b1e8; font-size:20px;'>üìä Step 5.1 |</span><span style='color:#6495ed;font-size:20px;'> Univariate Analysis üìä</span></b>
# <p style="background-color: #ff6347; padding: 10px; border-radius: 15px; color: #ffffff; font-size: 30px; font-family:Calibri; text-align:center;line-height: 1.5em; border-radius:100px 20px;">üìä 5.1.1 | Binary & Categorical Variables</p>

```python
def bar_char(column):
  bar_column = (data_train[column].value_counts(normalize=True) * 100).sort_values(ascending=False)
  fig = px.bar(
      bar_column,
      x=bar_column.index,
      y=bar_column.values,
      color=bar_column.index.astype(str),
      title=f'{column}',
      labels={'x': f"{column}" , 'y': 'Counts'},
      text=bar_column.apply(lambda x : f"{x:.2f}%"),
      template='plotly_dark'
  )

  fig.update_traces(
      textposition="outside",
      textfont={"size": 12, "family": "consolas"}
  )

  fig.update_layout(
      xaxis=dict(
          tickmode='linear',
          dtick=1
      )
  )

  return iplot(fig)
```

```python
# use function bart chart use in ploty:
cat_columns=[column for column in data_train.columns if data_train[column].nunique()<=4]
for column in cat_columns:
  bar_char(column)
```

# <p style="background-color: violet; padding: 1px; border-radius: 15px;color:white; font-size:30px; font-family:Calibri; text-align:center;line-height: 1.5em; border-radius:100px 20px;">üìä 5.1.2 | Numerical Variables</p>

```python
int_col=[column for column in data_train.columns if data_train[column].nunique()>8]
def kde_plot(int_columns):
  col_int=int_columns
  col=3
  rows=(len(col_int)+col-1)//col
  plt.figure(figsize=(15,16))
  for i,column in enumerate(col_int):
    plt.subplot(rows,col,i+1)
    sns.kdeplot(data=data_train, x=column, color='limegreen', fill=True)
  plt.suptitle('Distribution of numeric Variables', fontsize=40, y=1,color='orange')
  plt.tight_layout()
  plt.show()
kde_plot(int_col)
```

# <b><span style='color:#45b1e8; font-size:20px;'>üìä Step 5.2 |</span><span style='color:#6495ed;font-size:20px;'> Bivariate Analysis ü´Ä</span></b>
# <p style="background-color: #4169e1; padding: 1px; border-radius: 15px;color:white; font-size:30px; font-family:Calibri; text-align:center;line-height: 1.5em; border-radius:100px 20px;">üìä 5.2.1 | Binary & Categorical Variables</p>

```python
def cat_bet_target(cat_columns):
  col=3
  rows=(len(cat_columns)+col-1)//col
  plt.figure(figsize=(15, 12))
  for i,column in enumerate(cat_columns):
    plt.subplot(rows,col,i+1)
    sns.countplot(data=data_train,x=column,hue='Response')
    plt.title(f'{column} Distribution')
  plt.suptitle('Distribution of Categorical Variables by Target', fontsize=40, y=1,color='red')
  plt.tight_layout()
  plt.show()
cat_bet_target(cat_columns)
```

```python
def box_plot(int_col):
    #box plot:
    plt.figure(figsize=(18,13))
    num_col=3
    num_rows=(len(int_col)+num_col-1)//num_col
    for i ,column in enumerate(int_col):
      plt.subplot(num_rows,num_col,i+1)
      sns.boxplot(data=data_train,x='Response',y=column, palette='Set3')
    plt.suptitle('Distribution of numeric Variables',color='green',y=1,size=40)
    plt.tight_layout()
    plt.show()
box_plot(int_col)
```

```python
def Distribution_of_Numerical_Features(int_col):
    sns.set_palette("colorblind")  # Example: "colorblind", "Set2", "husl", etc.

    fig, axes = plt.subplots(len(int_col), 2, figsize=(15, 20))

    # Iterate over each selected column
    for i, column in enumerate(int_col):
        # Plot histogram without hue on the left
        sns.histplot(data=data_train, x=column, kde=True, bins=20, ax=axes[i, 0])
        axes[i, 0].set_title(f'Distribution of {column}')
        axes[i, 0].set_xlabel(column)
        axes[i, 0].set_ylabel('Frequency')

        # Plot box plot with hue on the right
        sns.boxplot(data=data_train, y=column, x='Response', ax=axes[i, 1])
        axes[i, 1].set_title(f'Distribution of {column} with Response Hue')
        axes[i, 1].set_xlabel(column)
        axes[i, 1].set_ylabel('Frequency')

    sns.despine()

    plt.tight_layout()

    plt.show()

Distribution_of_Numerical_Features(int_col)
```

```python
#Outliers_percentage data_train : 
def Outliers_percentage (data):
   percanteg=[]
   for column in int_col:
        q1=data[column].quantile(.25)
        q3=data[column].quantile(.75)
        iqr=q3-q1
        lower_bound=q1-1.5*iqr
        upper_bound=q3+1.5*iqr
        out_liers=data[(data[column]<lower_bound)|(upper_bound<data[column])]
        precanteg_out_liers=len(out_liers)/len(data)*100
        percanteg.append(precanteg_out_liers)
   plt.figure(figsize=(10,7))
   sns.barplot(x=int_col,y=percanteg)
   for i,v in enumerate(percanteg):
        plt.text(i,v+0.005,f"{v:.2f}",ha='center',va='bottom')
   plt.title('Percentage of Outliers in Each Column') 
   plt.xlabel('Columns')
   plt.ylabel('Percentage of Outliers')
   plt.xticks(rotation=45, ha='right')
   plt.tight_layout()
   plt.show()
```

```python
#Outliers_percentage data_train : 

Outliers_percentage(data_train)
```

```python
#Outliers_percentage data_test : 

Outliers_percentage(data_test)
```

```python
# handling_out_liers data _train
# Encoding Categorical Variables

def handling_out_liers(data):
    columns=int_col
    for column in columns:
      Q1=data[column].quantile(.25)
      Q3=data[column].quantile(.75)
      IQR=Q3-Q1
      lower_bounds=Q1-IQR*1.5
      upper_bounds=Q3+IQR*1.5
      data[column]=np.where(data[column]<lower_bounds,lower_bounds,data[column])
      data[column]=np.where(data[column]>upper_bounds,upper_bounds,data[column])
    return data
data_train=handling_out_liers(data_train)
data_test=handling_out_liers(data_test)
```

```python
# Data Preprocessing :
def Preprocessing(data):
    data['Gender']=data['Gender'].replace({'Male':1,'Female':0})
    data['Age']=data['Age'].astype(int)
    data['Region_Code']=data['Region_Code'].astype(int)
    data['Vehicle_Age']=data['Vehicle_Age'].map({'1-2 Year':0,'< 1 Year':1,'> 2 Years':2})
    data['Vehicle_Damage']=data['Vehicle_Damage'].map({'Yes':1,'No':0})
    data['Policy_Sales_Channel']=data['Policy_Sales_Channel'].astype(int)
    data['Annual_Premium']=data['Annual_Premium'].astype(int)
    data['Vintage']=data['Vintage'].astype(int)
Preprocessing(data_train)
Preprocessing(data_test)
```

```python
model_dict=dict()
def model(model,X_train,y_train,X_test,y_test,model_name):
    bulit_model=model
    bulit_model.fit(X_train,y_train)
    y_predict_train=bulit_model.predict(X_train)
    y_predict_test=bulit_model.predict(X_test)
    y_prdict_prop=bulit_model.predict_proba(X_test)[:,1]
    Auc=roc_auc_score(y_test,y_prdict_prop)
    acuuracy_data_train=accuracy_score(y_predict_train,y_train)
    acuuracy_data_test=accuracy_score(y_predict_test,y_test)
    print(f"the accuracy data train : {acuuracy_data_train}".title())
    print('*'*30)
    print(f"the accuracy data test : {acuuracy_data_test}".title())
    print('*'*50)
    print(f"The roc curv score : {Auc}".title())
    print('*'*50)
    print(classification_report(y_predict_train,y_train))
    print('*'*50)
    print(classification_report(y_predict_test,y_test))
    print('*'*50)
    confusion_matrix_train=confusion_matrix(y_predict_train,y_train)
    confusion_matrix_test=confusion_matrix(y_predict_test,y_test)
    plt.figure(figsize=(10,5))
    plt.subplot(1,2,1)
    sns.heatmap(confusion_matrix_train,annot=True,fmt='.1f',linewidths=2, linecolor='lightgrey')
    plt.title('confusion_matrix data train'.title())
    plt.subplot(1,2,2)
    sns.heatmap(confusion_matrix_test,annot=True,fmt='.1f',linewidths=2, linecolor='lightgrey')
    plt.title('confusion_matrix data test'.title())
    plt.show()
    model_dict[model_name]={'TRAIN ACCURACY':acuuracy_data_train,'TEST ACCURACY':acuuracy_data_test,'AUC':Auc}
```
